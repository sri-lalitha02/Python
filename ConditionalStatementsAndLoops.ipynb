```python
#Conditional Statements
# used to take decisions
# 1.if statement : The condition in the if block is true then the if block is executed
age = int(input("Enter the age : "))
if age > 18:
    print("You are eligiable for vote !")
# 2.else statement: The condition in the if block is false then the else block is executed
if age > 18:
    print("You are eligiable for vote !")
else:
    print("You are 'not' eligiable for vote !")
#single line if else statements
# Syntax : statement1 if condition else statement2
a,b = 10 , 4
print(a) if a > b else print(b)
# 3.elif statement : The condition in the if block is true then the elif block is executed
 # elif should not be written without a previous if
 # it’s not necessary to write an else statement after elif statment
cost_price = int(input("Enter cost price "))
selling_price = int(input("Enter selling price : "))
if cost_price > selling_price:
    print("Loss")
elif selling_price > cost_price:
    print("Profit")
else:
    print("No Profit And No Loss")
#  Nesting single line if else statements
# • Syntax: stmt1 if condition1 else stmt2 if condition2 else stmt3 if condition3 …. else stmt
print("Loss") if cost_price > selling_price else print("Profit") if selling_price > cost_price else print("No Profit And No Loss")
```

```python
# Loops
# Doing a process again and again based on a condition to achieve a task is looping
# 1.while loop :Block of statements will be executed as long as the condition is true
#To print even numbers from 1 to 10 
i = 1
while i <= 10:
    if(i%2 == 0):
        print(i , end=" ")
    i+=1
print()
# 2.for loop : Block of statements will be executed as long as the condition is true
# Syntak : for Variable in Sequence:
# Syntak : for Variable in (start , end+1 , step):
# In Python for loop always works on iterables (list , string  , range , tuple , set)
for ele in range (1 , 10 , 2):
    print(ele , end = " ")
print()
for var in range(10): # if you donot give start , step it take start as o and step as 1
    print(var , end = " ")
```

```python
# for loop on strings
str = "Python"
for ch in str:
    print(ch , end = " ")
print()
# for loop on lists
lst = ["Python" , "Java" , "Cpp" , "C" , 1 , 102.3 , [10 , 20 , 30]]
for ele in lst:
    print("element : " ,ele)
print()
tup = (10 , 20 , "Python" , "Java" , (1, 2 ,3))
for ele in tup:
    print(ele , end = " ")
```

```python
# Printing tables using while loop
i = 1 
while i <= 5:
    print(f"{5} x {i} = {5*i}")
    i = i+1
print()
# printing tables using for loop
num = 4  
for i in range (1 , 6 , 1):
    print("{} x {} = {}".format(num  , i , num*i))
```

```python
# range() method in python (built - in function) 
 # it is used to  generate integers fromone point to another point
# default values : start = 0 , step = 1
# 1. range(stop) 
 # generate integers from 0 to stop-1 and default values : start = 0 , step = 1
for i in range(11):
    print(i , end = " ")
print()
# 2. range(start , stop)
 #generates integers from start to stop-1 and default value : step = 1
for i in range(50 , 60):
    print(i , end = " ")
print()
# 3. range(start , stop , step)
 # generate values from start to stop with a difference of step 
for i in range(10 , 20 , 2):
    print(i , end = " ")
print()
## Generating numbers from backwards
 # A negative step value is required
 # syntax : for element in range(start , stop , -step)
for i in range(10 , 1 , -1):
   print(i , end = " ")
```

```python
# Truthiness of values in Python
# • In Python every value is having it’s own Truthvalue (True / False)

# • integers - Every integer, other than 0 will be treated as True
if 20 > 10:
    print("Hai")
else:
    print("Bye")
    
if -10:
    print("Hi")
    if 0:
        print("Hello")
    else:
        print("how are you ?")
else:
    print("False statement")

# • floating point values - Every point value, other than 0.0 will be treated as True
if 1.234:
    print("True")
else:
    print("False")
if 0.0000:
    print("Hi")
else:
    print("Bye")
    
# • Every sequence type with atleast one value in it will be treated as True
# • Empty sequences will be treated as False
if("hello"):
    print("Hai!")
else:
    print("Bye!")
if(''):
    print("Hello")
elif(' '):
    print("Hai")
else:
    print("Bye"
```

```python
# bool() --> returns either True or False
# • Every value can be checked for it’s Truthiness using bool() function
print("0.00 :" , bool(0.00))
print("-10 :", bool(-10))
print("10 :",bool(10))
print("This is python :" , bool("This is python"))
print("'' :",bool(''))
print("' ' :",bool(' '))
print("[] :",bool([]))
print("[0] :",bool([0]))
print("[10 , 20] :",bool([10,20])) 
print("() :",bool(()))
print("(1 , 2) :",bool((1,2)))
```
